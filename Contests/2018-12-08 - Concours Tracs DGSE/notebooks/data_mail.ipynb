```python
import pandas as pd


PATH_FILE_DATA_MAIL = "../data/mail_metadata_dataset.csv"
```

```python
df_mail = pd.read_csv(PATH_FILE_DATA_MAIL, sep=";")
df_mail["date"] = pd.to_datetime(df_mail["date"], format="%d-%m-%Y %H:%M")
df_mail.head()
```

```python
def get_receivers(sender, receiver):
    return pd.DataFrame(df_mail[df_mail["sender"].str.contains(sender, regex=False)]["receiver"][df_mail[df_mail["sender"].str.contains(sender, regex=False)]["receiver"].str.contains(receiver, regex=False)]).sort_values("receiver")
```

```python
from collections import Counter
s_ff = set(df_mail[df_mail["sender"].str.contains("felix.fachon", regex=False)]["receiver"].tolist())
s_gg = set(df_mail[df_mail["sender"].str.contains("georges-henri.gabrielli", regex=False)]["receiver"].tolist())
s_hh = set(df_mail[df_mail["sender"].str.contains("henry.huriolle", regex=False)]["receiver"].tolist())
s_ii = set(df_mail[df_mail["sender"].str.contains("ichak.illastrelli", regex=False)]["receiver"].tolist())
s_kk = set(df_mail[df_mail["sender"].str.contains("kevin.karchik", regex=False)]["receiver"].tolist())
(Counter(s_ff) + Counter(s_gg) + Counter(s_hh) + Counter(s_ii) + Counter(s_kk)).most_common(10)
```
